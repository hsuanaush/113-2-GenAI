# 宣告模組

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 畫函數圖形

```python
# 參數
petal = 7  # 花瓣數量

# 生成花朵形狀的參數方程
theta = np.linspace(0, 2 * np.pi, 1000)
r = 5 * (1 + np.sin(petal * theta))  # 調整花瓣數量
x = r * np.cos(theta)
y = r * np.sin(theta)

# 設定畫布
fig, ax = plt.subplots(figsize=(6, 6))
ax.fill(x, y, color='purple', alpha=0.6)  # 塗上紫色並設置透明度
ax.plot(x, y, color='darkviolet', linewidth=2)  # 邊框顏色加深

# 設定樣式
ax.set_aspect('equal')
ax.axis('off') # 不顯示座標
ax.set_title("Flower", fontsize=14) # 設定函數圖形標題

# 顯示圖形
plt.show()
```
